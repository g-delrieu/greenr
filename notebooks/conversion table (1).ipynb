```python
import pandas as pd
import re
data = pd.read_csv('Food_Weights.csv', sep = ";")
```

```python
data = data[data['Measure'].notna()]
```

```python
data = data.reset_index()
data = data.drop('index', axis = 1)
```

```python
quantity = []
unit = []

for i in range(len(data.Measure)):
    quantity.append(re.sub("[^0-9/ ]", "", data.Measure[i]))
    if re.search("[^0-9/]*$", data.Measure[i]).group(0) != '':
        unit.append(re.search("[^0-9/]*$", data.Measure[i]).group(0))
    else:
        unit.append('unit')
    
```

```python
def extract_quantity(quantity_str):
    if # value = nan replace 0
    print(re.sub(“[^0-9/ ]“, “”, quantity_str))
    return re.sub(“[^0-9/ ]“, “”, quantity_str)
data[‘quantity’] = data[‘Measure’].apply(extract_quantity)
```

```python
quantity = []
unit = []


for i in range(len(data.Measure)):
    #import pdb; pdb.set_trace()
    try:
        quantity.append(data.Measure[i].split()[0])
        unit.append(data.Measure[i].split()[1])
    except:
        quantity.append(data.Measure[i])
        unit.append(data.Measure[i])
```

```python
clean_data = pd.DataFrame(list(zip(data.Food, quantity, unit, data.Grams)), columns = ['Food', 'quantity', 'unit', 'weight'])
```

```python
clean_data.head(50)
```

```python
clean_data.unit.unique()
```

```python
clean_data = clean_data[clean_data['unit'] != ' lt.']
clean_data = clean_data[clean_data['unit'] != ' cup']
clean_data = clean_data[clean_data['unit'] != '-in. cube']
clean_data = clean_data[clean_data['unit'] != ' oz.']
clean_data = clean_data[clean_data['unit'] != '\xa0lb.']
clean_data = clean_data[clean_data['unit'] != ' cups']
clean_data = clean_data[clean_data['unit'] != '. Feb']
clean_data = clean_data[clean_data['unit'] != '-oz. can']
clean_data = clean_data[clean_data['unit'] != ' oz. or']
clean_data = clean_data[clean_data['unit'] != ' oz. or']
clean_data = clean_data[clean_data['unit'] != ' serving']
clean_data = clean_data[clean_data['unit'] != '-lb. loaf']
clean_data = clean_data[clean_data['unit'] != '-oz. bar']

```

```python
clean_data = clean_data.reset_index()
clean_data = clean_data.drop('index', axis = 1)
```

```python
clean_data.head(50)
```

```python
clean_data.quantity.unique()
```

```python
clean_quantity = []
for element in clean_data.quantity:
    if element == '1/4 ':
        clean_quantity.append(0.25)
    elif element == '1/2 ':
        clean_quantity.append(0.5)
    elif element == '':
        clean_quantity.append(0)
    elif element == ' ':
        clean_quantity.append(0)
    else:
        clean_quantity.append(float(element.strip()))
    
```

```python
clean_data['clean_quantity'] = clean_quantity
```

```python
clean_data = clean_data.drop('quantity', axis = 1)
```

```python
clean_data.weight.unique()
```

```python
clean_weight = []

for element in clean_data.weight:
    clean_weight.append(float(str(element).strip()))
```

```python
clean_data['clean_weight'] = clean_weight
```

```python
clean_data
```

```python
clean_data['unit_weight'] = clean_data['clean_weight']/clean_data['clean_quantity']
```
